```python
downsampled_df = pd.DataFrame(columns=crop_df.columns)
downsampled_df.shape
```

```python
# Loop through each unique crop in the dataset
for crop in crop_df['Crop'].unique():
    # Select rows for the current crop
    crop_rows = crop_df[crop_df['Crop'] == crop]

    # If there are more than 1000 rows for the current crop, downsample
    if len(crop_rows) > 1000:
        # Downsample to 1000 rows
        downsampled_rows = crop_rows.sample(n=1000, random_state=42)  # Use random_state for reproducibility

        # Append downsampled rows to the downsampled DataFrame
        downsampled_df = pd.concat([downsampled_df, downsampled_rows])
    else:
        downsampled_df = pd.concat([downsampled_df, crop_rows])

downsampled_df.reset_index(drop=True, inplace=True)
```

```python
# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Scale the features using Min-Max scaling
ms = MinMaxScaler()
x_train = ms.fit_transform(x_train)
x_test = ms.transform(x_test)
```

```python

# Standardize the scaled features

sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
#creating a dictionary of the following algorithums having key and value

models = {
    'Logistic Regression': LogisticRegression(),
    'Decision Tree': DecisionTreeClassifier(),
    'Gaussian Naive Bayes': GaussianNB(),
    'Support Vector Machine': SVC(),
    'K Nearest Neighbors': KNeighborsClassifier(),
    'Random Forest': RandomForestClassifier(),
    'Bagging Classifier': BaggingClassifier(),
    'Gradient Boosting': GradientBoostingClassifier(),
}

#now we will run a loop along this dictinoary
# Loop through the models
for name, model in models.items():
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)

    # Calculate and print accuracy, precision, recall, and F1-score
    accuracy = accuracy_score(y_test, y_pred)


    print(f"{name}:")
    print("Accuracy: {:.2f}%".format(accuracy * 100))

    print("\n")
```
